```python
%matplotlib inline
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
test = """Algoma.csv, Niagara.csv, Brant.csv, North Bay Parry Sound.csv, Chatham-Kent.csv, Northwestern.csv, City of Hamilton.csv, Ottawa.csv, Durham.csv, Peel.csv, Eastern ON.csv, Peterborough.csv, Grey Bruce.csv, Porcupine.csv, HKPR.csv, Renfrew.csv, Haldimand-Norfolk.csv, Simcoe Muskoka.csv, Halton.csv, Southwestern.csv, Hastings Prince Edward.csv, Sudbury.csv, Huron Perth.csv, Thunder Bay.csv, KFL&A.csv, Toronto.csv, LG&L.csv, WDG.csv, Lambton.csv, Waterloo.csv, Middlesex-London.csv, Windsor-Essex.csv"""
```

```python
csv_names = test.split(', ')
```

```python
for name in range(0, len(csv_names)):
    csv_names[name] = csv_names[name].replace('.csv', '')
```

```python
dataframes = []
```

```python
def convert_to_pd(names):
    for name in names:
        dataframes.append(pd.read_csv(f"School Municipalities/{name}.csv"))
    for name in range(0, len(dataframes)):
        try:
            dataframes[name].drop('Unnamed: 3', axis=1, inplace=True)
        except KeyError:
            pass
        dataframes[name].dropna(inplace=True)
        dataframes[name].set_index('collected_date', inplace=True)
```

```python
def plot_individual(num, kind="line", high=200, steps=10):
    print(csv_names[num])
    ax = dataframes[num].plot(kind=kind, figsize=(15,8), table=False, fontsize=13, linewidth=4.0, yticks=range(0, high, steps))
    ax = ax.legend(bbox_to_anchor=(1, 1.03), prop={'size':18})
    plt.xlabel("Date", fontsize=16)
    plt.ylabel("Cases", fontsize=16)
```

```python
def concat(num=[], every=True, student_only=False):
    for i in range(0, len(dataframes)):
        dataframes[i].rename(columns={dataframes[i].columns[0]: f"{csv_names[i]}_student_cases", 
                                      dataframes[i].columns[1]: f"{csv_names[i]}_total_cases"}, inplace=True)
    
    copy = dataframes
    if student_only == True:
        for i in range(0, len(copy)):
            copy[i].drop([copy[i].columns[1]], axis=1, inplace=True)
    
    if every == True:
        return pd.concat([x for x in copy], axis=1)
    else:
        temp = []
        for i in range(0, len(copy)):
            if i in num:
                temp.append(copy[i])
        return pd.concat([x for x in temp], axis=1)
        
```

```python
def plot_all(x, kind="line", high=200, steps=10):
    ax = x.plot(kind=kind, figsize=(15,8), table=False, fontsize=13, linewidth=4.0, yticks=range(0, high, steps))
    ax = ax.legend(bbox_to_anchor=(1, 1.03), prop={'size':18})
    plt.xlabel("Date", fontsize=16)
    plt.ylabel("Cases", fontsize=16)
```

```python
convert_to_pd(csv_names)
```

```python
plot_individual(1)
```

```python
y = concat([1, 4, 10], every=False)
```

```python
plot_all(y, high=100, steps=30)
```

```python
plot_individual(csv_names.index('Toronto'), high=800, steps=30)
```

```python
# (South East: toronto, durham, york) (West: labtom, chattam kent, windsor essex) (Middle: oxford, brant, waterloo) (East: ottawa, eastern ontario, leeds greenvile)
by_proximity = concat([csv_names.index('Peel'), csv_names.index('Toronto'),
                      csv_names.index('Chatham-Kent'), csv_names.index('Windsor-Essex'),
                      csv_names.index('Waterloo'), csv_names.index('Brant'),
                      csv_names.index('Ottawa'), csv_names.index('Eastern ON')], every=False, student_only=True)
```

```python
plot_all(by_proximity, high=700, steps=30)
```
